```python
import dask.dataframe as dd

column_names=["time","src_usr","dest_usr", \
                             "src_comp","dest_cmp","authentication_type", \
                             "logon_type","authentication_orientation","success_or_failure","day"]
ddf_train = dd.read_csv(
    '../lm-vol/LANL_train.csv', 
    names = column_names
)

ddf_test = dd.read_csv(
    '../lm-vol/LANL_test.csv', 
    names = column_names
)
```

```python
#doing this just to use Jacobs code

df= ddf_train

df = ddf_test

df = df.assign(src_user = df['src_usr'].str.split('@').map(lambda x: x[0]))
df = df.assign(dest_user = df['dest_usr'].str.split('@').map(lambda x: x[0]))

df = df[(df['src_usr'].str.contains('LOCAL')==False) & (df['src_usr'].str.contains('SYSTEM')==False) & (df['src_usr'].str.contains('ANONYMOUS')==False) & (df['src_usr'].str.contains('ADMINISTRARTOR')==False)]
df = df[df['src_user'].str[-1] != '$']

df = df[(df['dest_usr'].str.contains('LOCAL')==False) & (df['dest_usr'].str.contains('SYSTEM')==False) & (df['dest_usr'].str.contains('ANONYMOUS')==False) & (df['dest_usr'].str.contains('ADMINISTRARTOR')==False)]
df = df[df['dest_user'].str[-1] != '$']
```

```python
import pandas as pd
```

```python
unique_train = pd.read_csv('../lm-vol/LANL_train_unique.csv')
```

```python
unique_train.columns
```

```python
unique_train = unique_train.assign(src_user = unique_train['source user@domain'].str.split('@').map(lambda x: x[0]))
unique_train = unique_train.assign(dest_user = unique_train['destination user@domain'].str.split('@').map(lambda x: x[0]))

unique_train = unique_train[(unique_train['src_user'].str.contains('LOCAL')==False) & (unique_train['src_user'].str.contains('SYSTEM')==False) & (unique_train['src_user'].str.contains('ANONYMOUS')==False) & (unique_train['src_user'].str.contains('ADMINISTRARTOR')==False)]
unique_train = unique_train[unique_train['src_user'].str[-1] != '$']

unique_train = unique_train[(unique_train['dest_user'].str.contains('LOCAL')==False) & (unique_train['dest_user'].str.contains('SYSTEM')==False) & (unique_train['dest_user'].str.contains('ANONYMOUS')==False) & (unique_train['dest_user'].str.contains('ADMINISTRARTOR')==False)]
unique_train = unique_train[unique_train['dest_user'].str[-1] != '$']
```

```python
src_set = set(unique_train['source user@domain'].unique())
dest_set = set(unique_train['destination user@domain'].unique())
```

```python
all_train_set = src_set.union(dest_set)
```

```python
len(all_train_set)
```

```python
red_team_df = pd.read_csv('../lm-vol/redteam_complete.csv')
```

```python
red_src_set = set(red_team_df['source'].unique())
red_dest_set = set(red_team_df['destination'].unique())

all_red_set = red_src_set.union(red_dest_set)
```

```python
len(all_red_set)
```

```python
not_red_set = all_train_set.difference(all_red_set)
```

```python
len(not_red_set)
```

```python
import random 


sampled_2080 = random.sample(list(not_red_set), 2080)
```

```python
all_users = sampled_2080 + list(all_red_set)
```

```python
len(all_users)
```

```python
df = df[(df['src_usr'].isin(all_users)==True) & (df['dest_usr'].isin(all_users)==True) ]
```
